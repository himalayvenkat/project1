```python
# matplot lib is ploting
import matplotlib.pyplot as plt
```

```python
# pyplot is a sub package of matlib
```

```python
# multiple points
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# if you dont specify then line will be shown
```

```python
# two points
xpoints = [0,6]
ypoints = [0,250]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# plot has four arguments plot(x,y, format,)
# format ===> marker| line|color
# if i say marker = 'o'----> points
# if i say marker = 'x'> star points
# line ===> solid,dashed,dotted,dashdot
#color == > green, blue etc
```

```python
# two points

ypoints = [10,20,30,40]
plt.plot(ypoints,marker='*',ms = 20,color='g',mfc='b')
plt.show()
# by default it will take y points as 1,2,3,4....
# if we give one argument then it will show only take it as y points only
# ms means marker size
# mfc marker face color
```

```python
ypoints = [10,20,30,40]
plt.plot(ypoints, 'o:k')
plt.show()
```

```python
ypoints = [10,20,30,40]
plt.plot(ypoints, 'o-.k')
plt.show()
```

```python
ypoints = [10,20,30,40]
plt.plot(ypoints, 'o--k')
plt.show()
```

```python
# LIne width
ypoints = [10,20,30,40]
plt.plot(ypoints, 'o--k',linewidth='10')
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
x2 = [11,22,33,44,55]
y2 = [17,27,38,67,89]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
x2 = [11,22,33,44,55]
y2 = [17,27,38,67,89]
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
plt.plot(x1,y1)
plt.xlabel('days')
plt.ylabel('sales')
plt.title('Sales per day')
plt.show()
```

```python
# set fornt properties for title and label
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
plt.plot(x1,y1)
font1 = {'family':"serif",'color':"blue","size": 15}
font2 = {'family':"serif",'color':"green","size": 15}
font3 = {'family':"serif",'color':"red","size": 15}
plt.xlabel('days',fontdict = font1)
plt.ylabel('sales',fontdict = font2)
plt.title('Sales per day',fontdict = font3)
plt.show()
```
