```python
# two points
xpoints = [0,6]
ypoints = [0,250]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# plot has four arguments plot(x,y, format,)
# format ===> marker| line|color
# if i say marker = 'o'----> points
# if i say marker = 'x'> star points
# line ===> solid,dashed,dotted,dashdot
#color == > green, blue etc
```

```python
# two points

ypoints = [10,20,30,40]
plt.plot(ypoints,marker='*',ms = 20,color='g',mfc='b')
plt.show()
# by default it will take y points as 1,2,3,4....
# if we give one argument then it will show only take it as y points only
# ms means marker size
# mfc marker face color
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
plt.plot(x1,y1)
plt.xlabel('days')
plt.ylabel('sales')
plt.title('Sales per day')
plt.show()
```

```python
# set fornt properties for title and label
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
plt.plot(x1,y1)
font1 = {'family':"serif",'color':"blue","size": 15}
font2 = {'family':"serif",'color':"green","size": 15}
font3 = {'family':"serif",'color':"red","size": 15}
plt.xlabel('days',fontdict = font1,loc='right')
plt.ylabel('sales',fontdict = font2)
plt.title('Sales per day',fontdict = font3,loc='left')
plt.show()
# values= right.left,centre
```

```python
# to get the grid
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
plt.plot(x1,y1)
font1 = {'family':"serif",'color':"blue","size": 15}
font2 = {'family':"serif",'color':"green","size": 15}
font3 = {'family':"serif",'color':"red","size": 15}
plt.xlabel('days',fontdict = font1,loc='right')
plt.ylabel('sales',fontdict = font2)
plt.title('Sales per day',fontdict = font3,loc='left')
plt.grid() # shows grid lines
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
plt.plot(x1,y1)
font1 = {'family':"serif",'color':"blue","size": 15}
font2 = {'family':"serif",'color':"green","size": 15}
font3 = {'family':"serif",'color':"red","size": 15}
plt.xlabel('days',fontdict = font1,loc='right')
plt.ylabel('sales',fontdict = font2)
plt.title('Sales per day',fontdict = font3,loc='left')
plt.grid(color='g',linestyle="--",linewidth = 1,axis='x') # shows grid lines
plt.show()
```
